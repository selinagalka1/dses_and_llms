```python
from lxml import etree
import os

# === Reload trees ===
original_tree = etree.parse("french_plain.xml")
output_tree = etree.parse("annotated_with_claude.xml")
ns = {'tei': 'http://www.tei-c.org/ns/1.0'}

# === List of failed paragraph @xml:id values ===
failed_ids = ["P.200", "P.226", "P.317", "P.447", "P.460"]

# === Helper ===
def get_paragraph_by_id(tree, xml_id):
    return tree.xpath(f'//tei:p[@xml:id="{xml_id}"]', namespaces=ns)

# === Create logs directory ===
os.makedirs("logs", exist_ok=True)

# === Retry annotation ===
recovered = []
still_failed = []

for fid in failed_ids:
    print(f"ğŸ” Retrying annotation for: {fid}")
    original_p = get_paragraph_by_id(original_tree, fid)
    output_p = get_paragraph_by_id(output_tree, fid)

    if not original_p or not output_p:
        print(f"âš ï¸ Paragraph {fid} not found in one of the files.")
        still_failed.append(fid)
        continue

    paragraph_xml = etree.tostring(original_p[0], encoding="unicode")

    try:
        new_annotated_xml = annotate_whole_paragraph_with_claude(paragraph_xml)
        # Save raw Claude output for inspection
        with open(f"logs/raw_claude_output_{fid}.xml", "w", encoding="utf-8") as f:
            f.write(new_annotated_xml)

        try:
            new_element = etree.fromstring(new_annotated_xml.encode("utf-8"))
            parent = output_p[0].getparent()
            parent.replace(output_p[0], new_element)
            recovered.append(fid)
            print(f"âœ… Successfully updated paragraph {fid}")
        except etree.XMLSyntaxError as e:
            print(f"âš ï¸ Standard XML parsing failed for {fid}, trying recovery mode: {e}")
            try:
                parser = etree.XMLParser(recover=True)
                new_element = etree.fromstring(new_annotated_xml.encode("utf-8"), parser=parser)
                parent = output_p[0].getparent()
                parent.replace(output_p[0], new_element)
                recovered.append(fid)
                print(f"âœ… Recovered and updated paragraph {fid} (with recover=True)")
            except Exception as e:
                print(f"âŒ Still failed for {fid} (even in recovery mode): {e}")
                still_failed.append(fid)
    except Exception as e:
        print(f"âŒ Claude call failed for {fid}: {e}")
        still_failed.append(fid)

# === Save updated TEI ===
output_tree.write("annotated_with_claude_repaired.xml", encoding="utf-8", xml_declaration=True, pretty_print=True)
print("\nâœ… Final TEI saved to: annotated_with_claude_repaired.xml")

# === Report results ===
print("\nSummary:")
print(f"âœ… Recovered: {recovered}")
print(f"âŒ Still failed: {still_failed}")
```

```python
from lxml import etree
from collections import Counter
import csv

def extract_entities_from_paragraph(p_elem, tag_names):
    """Extract (tag, text) entities from paragraph, handling whitespace properly."""
    entities = []
    for tag in tag_names:
        for elem in p_elem.xpath(f".//tei:{tag}", namespaces={"tei": "http://www.tei-c.org/ns/1.0"}):
            text = elem.xpath("string()")
            if text:
                entities.append((tag, " ".join(text.strip().lower().split())))
    return entities


def compare_entities(gold_entities, model_entities):
    """Return TP, FP, FN and mismatched entities."""
    gold_counter = Counter(gold_entities)
    model_counter = Counter(model_entities)

    tp_entities = list((gold_counter & model_counter).elements())
    fp_entities = list((model_counter - gold_counter).elements())
    fn_entities = list((gold_counter - model_counter).elements())

    tp = len(tp_entities)
    fp = len(fp_entities)
    fn = len(fn_entities)

    return tp, fp, fn, tp_entities, fp_entities, fn_entities

def evaluate_tei_ner(gold_path, model_path, output_csv=None):
    tag_names = ["persName", "placeName"]

    gold_tree = etree.parse(gold_path)
    model_tree = etree.parse(model_path)

    gold_paragraphs = gold_tree.xpath("//tei:p", namespaces={"tei": "http://www.tei-c.org/ns/1.0"})
    model_paragraphs = model_tree.xpath("//tei:p", namespaces={"tei": "http://www.tei-c.org/ns/1.0"})

    if len(gold_paragraphs) != len(model_paragraphs):
        raise ValueError("Mismatch in number of paragraphs")

    total_tp = total_fp = total_fn = 0
    all_results = []

    for i, (gold_p, model_p) in enumerate(zip(gold_paragraphs, model_paragraphs), 1):
        gold_entities = extract_entities_from_paragraph(gold_p, tag_names)
        model_entities = extract_entities_from_paragraph(model_p, tag_names)

        tp, fp, fn, tp_e, fp_e, fn_e = compare_entities(gold_entities, model_entities)

        total_tp += tp
        total_fp += fp
        total_fn += fn

        print(f"\n--- Paragraph {i} ---")
        print(f"TP: {tp}, FP: {fp}, FN: {fn}")
        if fp_e:
            print("False Positives (extra in model):")
            for e in fp_e:
                print(f"  {e[0]}: {e[1]}")
        if fn_e:
            print("False Negatives (missing in model):")
            for e in fn_e:
                print(f"  {e[0]}: {e[1]}")

        all_results.append({
            "paragraph": i,
            "tp": tp,
            "fp": fp,
            "fn": fn,
            "false_positives": "; ".join([f"{tag}:{text}" for tag, text in fp_e]),
            "false_negatives": "; ".join([f"{tag}:{text}" for tag, text in fn_e]),
        })

    # Overall scores
    precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) else 0
    recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

    print("\n=== Overall Evaluation ===")
    print(f"Total True Positives: {total_tp}")
    print(f"Total False Positives: {total_fp}")
    print(f"Total False Negatives: {total_fn}")
    print(f"Precision: {precision:.3f}")
    print(f"Recall:    {recall:.3f}")
    print(f"F1 Score:  {f1:.3f}")

    if output_csv:
        keys = all_results[0].keys()
        with open(output_csv, "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=keys)
            writer.writeheader()
            writer.writerows(all_results)
        print(f"\nDetailed results saved to: {output_csv}")

# === USAGE ===
# Replace these with your file paths:
evaluate_tei_ner("french_gold.xml", "annotated_with_claude.xml", output_csv="schwerin_sample_20p_GPT/evaluation_results/ner_evaluation_results_gpt.csv")
```

```python
from lxml import etree
from xmldiff import main
from xmldiff.formatting import DiffFormatter, WS_BOTH

# Define the TEI namespace
NS = {"tei": "http://www.tei-c.org/ns/1.0"}

def clean_tei(file_path):
    """
    Parses and cleans a TEI XML file by removing <persName> and <placeName> tags
    (while preserving their text content) and deleting 'ref' and 'wit' attributes.
    """
    parser = etree.XMLParser(remove_blank_text=True)
    tree = etree.parse(file_path, parser)
    root = tree.getroot()

    # Remove <persName> and <placeName> tags but keep their text content
    for tag in root.xpath(".//tei:persName | .//tei:placeName", namespaces=NS):
        parent = tag.getparent()
        if tag.text:
            # Append the text to the tail of the previous element or the parent text
            if parent is not None:
                if parent.text:
                    parent.text += tag.text
                else:
                    parent.text = tag.text
        if tag.tail:
            # Append the tail text to the next sibling or parent
            if tag.getnext() is not None:
                if tag.getnext().tail:
                    tag.getnext().tail = tag.tail + tag.getnext().tail
                else:
                    tag.getnext().tail = tag.tail
            elif parent is not None:
                if parent.tail:
                    parent.tail += tag.tail
                else:
                    parent.tail = tag.tail
        parent.remove(tag)

    # Remove 'ref' and 'wit' attributes
    for elem in root.xpath(".//*[@ref] | .//*[@wit]", namespaces=NS):
        elem.attrib.pop("ref", None)
        elem.attrib.pop("wit", None)

    return tree

def compare_tei_files(file1, file2):
    """
    Compares two TEI XML files after cleaning and prints the differences.
    """
    # Clean the TEI files
    tree1 = clean_tei(file1)
    tree2 = clean_tei(file2)

    # Convert the cleaned trees to strings
    xml1 = etree.tostring(tree1, encoding='unicode')
    xml2 = etree.tostring(tree2, encoding='unicode')

    # Set up the formatter with whitespace normalization
    formatter = DiffFormatter(normalize=WS_BOTH)

    # Compute the differences
    diff = main.diff_texts(xml1, xml2, formatter=formatter)

    # Output the results
    if not diff.strip():
        print("âœ… No differences found â€” TEI files are equivalent after cleaning.")
    else:
        print("âŒ Differences found:")
        print(diff)

# Example usage
file1 = "input/gold.xml"
file2 = "output/claude/prompt3.xml"
compare_tei_files(file1, file2)
```
